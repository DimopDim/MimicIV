```python
"""
Γεμίζω τον πίνακα με τις μετρήσεις του chartevet 8ώρου.
"""
```

```python
import pandas as pd
```

```python
df_chart_tbl = pd.read_csv('D:\MSc Aegean\Διατριβή\MIMIC_IV\mimic-iv-1.0\exports\Test.csv', index_col=False)
df_grouped_cerebral_infarction = pd.read_csv('D:\MSc Aegean\Διατριβή\MIMIC_IV\mimic-iv-1.0\exports\TestII.csv', index_col=False)
```

```python
# Δημιουργώ εκ νέου την στήλη concat χωρίς κενά ανάμεσα στα στοιχεία.
# Θα χρησιμοποιήσω αυτή τη στήλη σαν index και τα κενά δημιουργούν πρόβλημα.
df_chart_tbl["concat"] = df_chart_tbl["subject_id"].astype(str) + df_chart_tbl["hadm_id"].astype(str) + df_chart_tbl["date"].astype(str) + df_chart_tbl["time"].astype(str)
```

```python
df_chart_tbl['concat'] = df_chart_tbl['concat'].str.replace('-','')
df_chart_tbl['concat'] = df_chart_tbl['concat'].str.replace(':','')
```

```python
# Ορίζω την στήλη concat ως index των γραμμών
df_chart_tbl = df_chart_tbl.set_index('concat')
```

```python
df_grouped_cerebral_infarction['charttime']= pd.to_datetime(df_grouped_cerebral_infarction['charttime'])
```

```python
df_grouped_cerebral_infarction
```

```python
df_grouped_cerebral_infarction = df_grouped_cerebral_infarction.groupby(['subject_id', 'hadm_id', 'date', 'time', 'itemid', pd.Grouper(key='charttime', freq='1D')], dropna=False).agg({'values': ['mean']})
df_grouped_cerebral_infarction.columns = ['Mean_Chart']
df_grouped_cerebral_infarction = df_grouped_cerebral_infarction.reset_index()
df_grouped_cerebral_infarction
```

```python
# Δημιουργώ την στήλη concat και στον πίνακα που περιλαμβάνει τις μετρήσεις για να είναι το κοινώ σημείο ανάμεσα στους δυο πίνακες
df_grouped_cerebral_infarction["concat"] = df_grouped_cerebral_infarction["subject_id"].astype(str) + df_grouped_cerebral_infarction["hadm_id"].astype(str) + df_grouped_cerebral_infarction["date"].astype(str) + df_grouped_cerebral_infarction["time"].astype(str)
```

```python
df_grouped_cerebral_infarction
```

```python
df_grouped_cerebral_infarction['concat'] = df_grouped_cerebral_infarction['concat'].str.replace('-','')
df_grouped_cerebral_infarction['concat'] = df_grouped_cerebral_infarction['concat'].str.replace(':','')
```

```python
# Μεταφέρω την τελευταία στήλη πρώτη
cols = list(df_grouped_cerebral_infarction.columns)
cols = [cols[-1]] + cols[:-1]
df_grouped_cerebral_infarction = df_grouped_cerebral_infarction[cols]
```

```python
df_grouped_cerebral_infarction
```

```python
for row in df_grouped_cerebral_infarction.itertuples():
    con = row.concat
    item = row.itemid
    value = row.Mean_Chart
    #print('{}'.format(con), '{}'.format(item), '{}'.format(value))
    df_chart_tbl.at['{}'.format(con), '{}'.format(item)] = value
```
