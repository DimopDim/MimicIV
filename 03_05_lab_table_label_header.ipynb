```python
#Εισαγωγή βιβλιοθήκης Panda
import pandas as pd
```

```python
df_data_labevents = pd.read_csv('D:\MSc Aegean\Διατριβή\MIMIC_IV\mimic-iv-1.0\hosp\d_labitems.csv.gz')
df_item_header = pd.read_csv('D:\MSc Aegean\Διατριβή\MIMIC_IV\mimic-iv-1.0\exports\o15_fill_lab_itemid_header_tbl.csv')
```

```python
df_data_labevents = df_data_labevents.set_index('itemid')
```

```python
# Ενοποίηση τιμών σε κελιά
df_data_labevents["header"] = (df_data_labevents["label"].astype(str)
                                                + " ("
                                                + df_data_labevents["fluid"].astype(str)
                                                + ")")
```

```python
# Ελέγχω τις κεφαλίδες και αν δεν ανήκουν στις
# τέσσερις πρώτες τιμές κάνω τις αντικατάσταση.
for col in df_item_header.columns:
    if ((col != 'subject_id') and (col != 'hadm_id') and (col != 'Department') and (col != 'date') and (col != 'time')):
        temp = (df_data_labevents.loc[int(col), 'header'])
        df_data_labevents.rename(columns={'{}'.format(col): '{}'.format(temp)}, inplace=True)
```
